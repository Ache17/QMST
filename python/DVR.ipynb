```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hh = 0.0005
re=1.4011
mu=918.58
deltr = 0.8
rmin = re - deltr
rmax = re + deltr

M = np.floor( 2 * deltr / hh) + 1
h = (rmax - rmin) / (M - 1)
r = np.arange(rmin, rmax+ h, h)

k4=2.8417

v = 0.5 * k4 * (r - re)**4
pemat = np.diag(v)

kemat = 16 * np.diag( np.ones(int(M)-1),1)
kemat = kemat - np.diag(np.ones(int(M) - 2), 2)
kemat = kemat + np.transpose(kemat)
kemat = kemat - 30 * np.diag(np.ones(int(M)), 0)
kemat = kemat / (12 * h*h)
kemat = -kemat/ (2 * mu)



hmat  = kemat + pemat 

eigenvalues, eigenvectors = np.linalg.eig(hmat)

#n = 1 / np.sqrt(h)
#eigenvectors = eigenvectors * n
energies = np.diag(np.sort(eigenvalues)[::-1])
print(energies[0][0], energies[1][1])
```

```python
plt.figure(figsize=(16, 16))
plt.plot(r, eigenvectors[0])
plt.show()
```
